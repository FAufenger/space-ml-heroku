## Read in original_data

```python
#Dependencies
import pandas as pd
import numpy as np
import os
import time

#File load
impacts = "../Resources/original_data/impacts.csv"
orbits = "../Resources/original_data/orbits.csv"
Stars = "../Resources/original_data/Stars.csv"
UFO = "../Resources/original_data/scrubbed.csv"

#Read file and store into pandas data frame
impacts_df = pd.read_csv(impacts, sep = ',')
orbits_df = pd.read_csv(orbits, sep = ',')
Stars_df = pd.read_csv(Stars, sep = ',')
ufo_df = pd.read_csv(UFO, sep = ',')
```

# Clean and inspect data

### asteroid impacts dataframe

```python
print(f'The length of this df is: {len(impacts_df)}')
```

```python
# Clean columns
impacts_df.columns = ['object_name','period_start','period_end','possible_impacts', 
              'cumulative_impact_probability', 'asteroid_velocity', 'asteroid_magnitude', 
              'asteroid_diameter_km', 'cumulative_palermo_scale', 'maximum_palermo_scale', 
              'maximum_torino_scale']
```

```python
impacts_df = impacts_df.drop(columns=['maximum_torino_scale']);
```

### UFO dataframe

```python
print(f'The length of this df is: {len(ufo_df)}')
# ufo_df.head(4)
```

```python
ufo_df.columns = ['datetime','city','state','country','shape','duration_seconds','duration_hours_min','comments','date_posted','latitude','longitude']
```

```python
## Index 43782 has latitude of 33q.200088, 
## have to fix or latitude to be read properly
ufo_df.loc[(ufo_df.latitude == '33q.200088'), 'latitude'] = '33.200088'
```

```python
# date_posted to date time
ufo_df['date_posted'] = pd.to_datetime(ufo_df.date_posted)
# ufo_df.head(4)
```

```python
# Convert date time to yyyy-mm-dd  hh:mm:ss with split value columns
dateTime = ufo_df["datetime"].str.split(" ", n = 1, expand = True)
dateTime.columns = ['date', 'time']

dateTime['date'] = pd.to_datetime(dateTime.date)

print(f'the length of the new df is: {len(dateTime)}')
dateTime.head(4)
```

```python
# Add new columns and drop old
ufo_df["date_occured"]= dateTime["date"]
ufo_df["time_occured"]= dateTime["time"]

ufo_df.drop(columns =["datetime"], inplace= True)
```

```python
# Rearrange columns
ufo_df = ufo_df[['date_occured','time_occured','city','state','country',
                 'shape','duration_seconds','duration_hours_min','comments',
                 'date_posted','latitude','longitude']]
```

```python
ufo_df
```

```python
# ufo_df.loc['1940-01-01':'1989-03-12']
# ufo_df.loc[(ufo_df.date_occured == ('1940-01-01':'1989-03-12')
# Filter data between two dates
ufo_reduced_df = ufo_df.loc[(ufo_df['date_occured'] >= '1940-01-01')
                     & (ufo_df['date_occured'] <= '1989-12-31')]
```

```python
print(len(ufo_reduced_df))
ufo_reduced_df.head(5)
```
